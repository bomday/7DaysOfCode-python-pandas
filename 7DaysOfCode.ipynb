## Python Pandas: Limpeza de Dados e Manipulações de Dataframe

### Adicionando Nova Coluna

```python
cdu = []

for localizacao in emprestimos_completo['localizacao']:
  if localizacao < 100:
    cdu.append('Generalidades. Ciência e conhecimento.')
  elif 100 <= localizacao < 200:
    cdu.append('Filosofia e psicologia.')
  elif 200 <= localizacao < 300:
    cdu.append('Religião.')
  elif 300 <= localizacao < 400:
    cdu.append('Ciências sociais.')
  elif 400 <= localizacao < 500:
    cdu.append('Classe vaga. Provisoriamente não ocupada.')
  elif 500 <= localizacao < 600:
    cdu.append('Matemática e ciências naturais.')
  elif 600 <= localizacao < 700:
    cdu.append('Ciências aplicadas.')
  elif 700 <= localizacao < 800:
    cdu.append('Belas artes.')
  elif 800 <= localizacao < 900:
    cdu.append('Linguagem. Língua. Linguística.')
  elif 900 <= localizacao < 1000:
    cdu.append('Geografia. Biografia. História.')

emprestimos_completo['CDU'] = cdu
emprestimos_completo
```

### Excluindo Colunas

```python
emprestimos_completo = emprestimos_completo.drop(['registro_sistema'], axis=1)
emprestimos_completo.head()
```

```python
# Convertendo os valores da coluna 'matricula_ou_siape' para string
emprestimos_completo['matricula_ou_siape'] = emprestimos_completo['matricula_ou_siape'].apply(str)

# Convertendo as colunas 'data_renovacao', 'data_emprestimo' e 'data_devolucao' para datetime
emprestimos_completo['data_renovacao'] = pd.to_datetime(emprestimos_completo['data_renovacao'])
emprestimos_completo['data_emprestimo'] = pd.to_datetime(emprestimos_completo['data_emprestimo'])
emprestimos_completo['data_devolucao'] = pd.to_datetime(emprestimos_completo['data_devolucao'])

emprestimos_completo.head()
```

### Análise Quantidade de Exemplares por Ano

```python
# Cria cópia de dataframe com valor da coluna 'data_emprestimo'
emprestimos_por_ano = emprestimos_completo.copy().loc[:, 'data_emprestimo'].dt.year

# Faz agrupamento de quantidade de empréstimos por ano
emprestimos_por_ano = emprestimos_por_ano.value_counts().sort_index(ascending=False).to_frame('total')
emprestimos_por_ano.index.rename('ano', inplace=True)
emprestimos_por_ano
```

```python
# Plotar o gráfico de linhas
plt.figure(figsize=(8, 4))
plt.plot(emprestimos_por_ano.index, emprestimos_por_ano['total'], marker='o', linestyle='-', color='m', label='Valor')

# Adicionar título e rótulos ao gráfico
plt.title('Gráfico 1: Quantidade de Empréstimos por Ano')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Empréstimos')

# Apresentar o gráfico
plt.show()
```

*   Houve crescimento na quantidade de empréstimos anuais no período entre 2010 e 2013;
*   No ano de 2014, verificou-se uma queda em relação ao ano anterior;
*   No período de 2014 a 2017, a quantidade de empréstimos anuais não variou significativamente;
*   Após o ano de 2018, houve queda na quantidade de empréstimos anuais, principalmente, no ano de 2020;
*   O ano de 2013 obteve mais destaque na quantidade de empréstimos anuais.







### Análise Quantidade de Exemplares por Mês

```python
# Cria cópia de dataframe com valor da coluna 'data_emprestimo'
emprestimos_por_mes = emprestimos_completo.copy().loc[:, 'data_emprestimo'].dt.month

# Faz agrupamento de quantidade de empréstimos por ano
emprestimos_por_mes = emprestimos_por_mes.value_counts().sort_index(ascending=True).to_frame('total')
emprestimos_por_mes.index.rename('mes', inplace=True)

# Modificando os números por nomes dos meses
dicionario_meses = {1:'Jan',2:'Fev',3:'Mar',4:'Abr',
                    5:'Mai',6:'Jun',7:'Jul',8:'Ago',
                    9:'Set',10:'Out',11:'Nov',12:'Dez'}

emprestimos_por_mes.index = emprestimos_por_mes.index.map(dicionario_meses)
emprestimos_por_mes
```

```python
# Plotar o gráfico de linhas
plt.figure(figsize=(8, 4))
plt.plot(emprestimos_por_mes.index, emprestimos_por_mes['total'], marker='o', linestyle='-', color='m', label='Valor')

# Adicionar título e rótulos ao gráfico
plt.title('Gráfico 2: Quantidade de Empréstimos por Mês')
plt.xlabel('Mês')
plt.ylabel('Quantidade de Empréstimos')

# Apresentar o gráfico
plt.show()
```

```python
# Criar o gráfico de barras
plt.figure(figsize=(8, 4))  # Definir o tamanho do gráfico
plt.bar(emprestimos_por_mes.index, emprestimos_por_mes['total'], color='skyblue')

# Personalizar o gráfico
plt.title('Gráfico 3: Quantidade de Empréstimos por Mês')
plt.xlabel('Mês')
plt.ylabel('Quantidade de Empréstimos')

# Apresentar o gráfico
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

*   Os meses com maiores quantidades de empréstimos mensais foram: **Março** e **Agosto**;
*   Os meses mais indicados para férias dos trabalhadores são **Janeiro** e **Dezembro**, pela baixa demanda de quantidades de empréstimos mensais;
*   Recomenda-se manter as atividades que não sejam de atendimento ao usuário para o período de **Dezembro a Janeiro** e **Junho a Julho**, já que os estudantes estarão, comumente, de férias e não terá uma demanda tão grande para atendimento presencial.







### Análise Quantidade de Exemplares por Hora

```python
# Cria cópia de dataframe com valor da coluna 'data_emprestimo'
emprestimos_por_hora = emprestimos_completo.copy().loc[:, 'data_emprestimo'].dt.hour

# Faz agrupamento de quantidade de empréstimos por ano
emprestimos_por_hora = emprestimos_por_hora.value_counts().sort_index(ascending=True).to_frame('total')
emprestimos_por_hora.index.rename('hora', inplace=True)
emprestimos_por_hora
```

```python
# Criar o gráfico de barras
plt.figure(figsize=(8, 4))
plt.bar(emprestimos_por_hora.index, emprestimos_por_hora['total'], color='skyblue')

# Personalizar o gráfico
plt.title('Gráfico 4: Quantidade de Empréstimos por Hora')
plt.xlabel('Hora')
plt.ylabel('Quantidade de Empréstimos')

# Apresentar o gráfico
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Adicionar grade no eixo Y
plt.tight_layout()  # Ajustar o layout do gráfico
plt.show()
```

*   É possível identificar que os períodos das 10 h e 16 h são os que fornecem mais empréstimos;
*   Períodos do começo da manhã (entre 7h e 9h) e da noite (entre 19h e 22h) podem ser os melhores horários para que se realizem outras atividades não relacionadas ao atendimento de cliente;

*   Entre 12h e 13h são momentos de demandas menores. Porém, por ser um horário onde a maioria dos funcionários também estarão em almoço, pode não ser tão interessante alocar outras demandas.

### Função Tabelas

```python
def tabela(var):
  emprestimos_por_var = emprestimos_completo[var].value_counts().to_frame('quantidade')
  emprestimos_por_var['percentual'] = round((emprestimos_por_var.quantidade/emprestimos_por_var.quantidade.sum())*100,1).sort_index(ascending=True)
  emprestimos_por_var.index.rename(None, inplace=True)

  return emprestimos_por_var
```

### Percentual de Empréstimos por Tipo de Vínculo de Usuário

```python
emprestimos_por_vinculo = tabela('tipo_vinculo_usuario')
emprestimos_por_vinculo
```

### Percentual de Empréstimos por Coleção

```python
emprestimos_por_colecao = tabela('colecao')
emprestimos_por_colecao
```

### Percentual de Empréstimos por Biblioteca

```python
emprestimos_por_biblioteca = tabela('biblioteca')
emprestimos_por_biblioteca
```

### Percentual de Empréstimos por CDU - Classificação Decimal Universal

```python
emprestimos_por_cdu = tabela('CDU')
emprestimos_por_cdu
```
